Literal a,b,c

```python
import numpy as np

# Definir la matriz de coeficientes
B = np.array([
    [1, 0, -1],
    [-1/2, 1, -1/4],
    [1, -1/2, 1]
])

# Vector de términos independientes
b = np.array([0.2, -1.425, 2])

# Condiciones iniciales para el método de Gauss-Seidel
x = np.zeros(3)

def estrictamente_dominante(A):
   
    n = len(A)
    for i in range(n):
        suma_fila = sum(abs(A[i][j]) for j in range(n) if j != i)
        if abs(A[i][i]) <= suma_fila:
            return False
    return True

def gauss_seidel(A, b, x, tol, max_iter):

    n = len(b)
    for k in range(max_iter):
        x_old = x.copy()
        for i in range(n):
            sum1 = sum(A[i][j] * x[j] for j in range(i))
            sum2 = sum(A[i][j] * x_old[j] for j in range(i + 1, n))
            x[i] = (b[i] - sum1 - sum2) / A[i][i]

        # Criterio de parada
        if np.linalg.norm(x - x_old, ord=np.inf) < tol:
            return x, k + 1, "Convergencia Encontrada."
    return x, max_iter, "No converge."

# Tolerancia y máximo de iteraciones
tolerance = 10**22
max_iterations = 300

# **Literal a: Verificación de diagonal estrictamente dominante**
diagonal_dominante = estrictamente_dominante(B)
print(f"Literal a: La matriz es diagonal estrictamente dominante: {diagonal_dominante}")


# **Literal b: Aplicación del método de Gauss-Seidel**
solution, iterations, status = gauss_seidel(B, b, x, tolerance, max_iterations)
print("Literal b: Resultado del método de Gauss-Seidel:")
print(f"Solución aproximada: {solution}")
print(f"Iteraciones realizadas: {iterations}")
print(f"Estado: {status}")

# **Literal c: Cambio en el sistema con una nueva matriz**
B_nueva = np.array([
    [1, 0, -2],
    [-1/2, 1, -1/4],
    [1, -1/2, 1]
])

print("\nLitearl c: Analizando el nuevo sistema con matriz modificada:")
new_diagonal_dominante = estrictamente_dominante(B_nueva)
print(f"La nueva matriz es diagonal estrictamente dominante: {new_diagonal_dominante}")

# Aplicar Gauss-Seidel al nuevo sistema
solution_nueva, iterations_nueva, status_nueva = gauss_seidel(B_nueva, b, x, tolerance, max_iterations)
print("Resultado del método de Gauss-Seidel para la nueva matriz:")
print(f"Solución aproximada: {solution_nueva}")
print(f"Iteraciones realizadas: {iterations_nueva}")
print(f"Estado: {status_nueva}")
```

```python
import numpy as np

# Definir la matriz de coeficientes
B = np.array([
    [1, 0, -1],
    [-1/2, 1, -1/4],
    [1, -1/2, 1]
])

# Vector de términos independientes
b = np.array([0.2, -1.425, 2])

# Condiciones iniciales para el método de Gauss-Jacobi
x = np.zeros(3)

def estrictamente_dominante(A):
    n = len(A)
    for i in range(n):
        suma_fila = sum(abs(A[i][j]) for j in range(n) if j != i)
        if abs(A[i][i]) <= suma_fila:
            return False
    return True

# Gauss-Jacobi
def gauss_jacobi(A, b, x, tol, max_iter):
    n = len(b)
    for k in range(max_iter):
        x_old = x.copy()
        for i in range(n):
            suma = sum(A[i][j] * x_old[j] for j in range(n) if j != i)
            x[i] = (b[i] - suma) / A[i][i]

        # Criterio de parada
        if np.linalg.norm(x - x_old, ord=np.inf) < tol:
            return x, k + 1, "Convergencia Encontrada."
    return x, max_iter, "No converge."

# Tolerancia y máximo de iteraciones
tolerance = 1e-10
max_iterations = 300

# **Literal a: Verificación de diagonal estrictamente dominante**
diagonal_dominante = estrictamente_dominante(B)
print(f"Literal a: La matriz es diagonal estrictamente dominante: {diagonal_dominante}")

# **Aplicación del método de Gauss-Jacobi**
solution_jacobi, iterations_jacobi, status_jacobi = gauss_jacobi(B, b, x.copy(), tolerance, max_iterations)
print("Resultado del método de Gauss-Jacobi:")
print(f"Solución aproximada: {solution_jacobi}")
print(f"Iteraciones realizadas: {iterations_jacobi}")
print(f"Estado: {status_jacobi}")

# **Literal c: Cambio en el sistema con una nueva matriz**
B_nueva = np.array([
    [1, 0, -2],
    [-1/2, 1, -1/4],
    [1, -1/2, 1]
])

print("\nLiteral c: Analizando el nuevo sistema con matriz modificada:")
new_diagonal_dominante = estrictamente_dominante(B_nueva)
print(f"La nueva matriz es diagonal estrictamente dominante: {new_diagonal_dominante}")

# Aplicar Gauss-Jacobi al nuevo sistema
solution_nueva_jacobi, iterations_nueva_jacobi, status_nueva_jacobi = gauss_jacobi(B_nueva, b, x.copy(), tolerance, max_iterations)
print("Resultado del método de Gauss-Jacobi para la nueva matriz:")
print(f"Solución aproximada: {solution_nueva_jacobi}")
print(f"Iteraciones realizadas: {iterations_nueva_jacobi}")
print(f"Estado: {status_nueva_jacobi}")
```
